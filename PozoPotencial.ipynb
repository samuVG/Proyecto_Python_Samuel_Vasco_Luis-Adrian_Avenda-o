```python
import Pozo as p
import numpy as np

N=1000
hbar2=(4.135e-15/(2*np.pi))**2
z =np.linspace(-15e-10,15e-10,N)
h=(z[1]-z[0])/2
a=10e-10
m=0.51e6/(3e8)**2
```

# **Pozo de Potencial de paredes finitas** 

$V(z)=\begin{cases}
 0& \text{ if } |z|\leq \frac{a}{2} \\ 
  v_0& \text{ if } |z|>\frac{a}{2}
\end{cases}$

```python
v0=15
A=p.PozoDePotencial(z,N,hbar2,h,a,m,v0)

def Potencial(z,v0):
    if (z<(-a/2) or z>(a/2)):
        return v0
    else: 
        return 0
    
# MATRIZ HAMILTONIANA
H=p.PozoDePotencial.MatrizHamiltoniana(A,Potencial) 

# 4 primeros AUTOVALORES
autovalores=p.PozoDePotencial.Autovalores(A,H)

# 4 primeras AUTOFUNCIONES
autofunciones=p.PozoDePotencial.Autofunciones(A,H)

# DENSIDADES DE PROBABILIDAD
Desidad_Probabilidad=p.PozoDePotencial.DensidadProbabilidad(A,H)

# AUTOFUNCION DEPENDIENTE DEL TIEMPO
Autofuncion_Temporal= lambda t: p.PozoDePotencial.AutofuncionTemporal(A,H,t)

# Grafica de las autofunciones
p.PozoDePotencial.GraficadorAutofunciones(A,H,Potencial)

# Grafica de las densidades de probabilidad
p.PozoDePotencial.GraficadorDensidadProbabilidad(A,H,Potencial)

# Grafica de la densidad de probabilidad temporal para diferentes tiempos
p.PozoDePotencial.GraficadorAutofuncionTemporal(A,H,Potencial)

print("NOTA: en la grafica de las autofunciones se grafica V(z) vs z, y sobre esta las autofunciones centradas en los respectivos autovalores para mostrar que estos corresponden a los 4 primeros estados ligados. Luego se hace ZOOM en los autovalores para mostrar la forma de la autofunción de onda en dicho autovalor o autoestado.")
```
